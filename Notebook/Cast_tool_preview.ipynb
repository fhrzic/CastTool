## Load libraries and upload the data

Run the cell below by selecting it and pressing "ctrl+enter." The prompt asks for data selection, select **"Data.zip"** archive, and start the upload.

```python
# Import libraries
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from IPython.display import display
import cv2
import pathlib
import matplotlib.pyplot as plt
import IPython.display as display
import zipfile
import io
from google.colab import files

# Upload data
print("Select data for the upload:")
uploaded = files.upload()
data = zipfile.ZipFile(io.BytesIO(uploaded['Data.zip']), 'r')
data.extractall()
data.printdir()

# Real and generated images directories
real_image_dir = './Data/real/'
generated_image_dir = './Data/generated/'
```

## Run simulation

The alpha and image selection widgets will be enabled by running the cell below ("ctrl+enter"). Feel free to experiment.

```python
# Get images names
def get_images_list(dir):
    data_root = pathlib.Path(dir)
    # Get images paths
    # Get image names
    image_paths = list(data_root.glob('./*.png'))    
    all_image_paths = [str(path) for path in image_paths]
    image_names = [name.split('/')[-1] for name in all_image_paths]
    return image_names

# Draw
def f(alpha, image):
    real_image = real_image_dir+image
    generated_image = generated_image_dir+image
    
    r_img = cv2.imread(real_image, 0)
    g_img = cv2.imread(generated_image, 0)
     
    new_img = (1-alpha)*r_img + alpha*(g_img)
    
    %matplotlib inline
    plt.figure(figsize = (10,10))
    imgplot = plt.imshow(new_img, interpolation='nearest', cmap='gray')


# Interact    
image_names= get_images_list(real_image_dir)
image_names.sort(key = lambda x: int(x.split('.')[0]))
w = widgets.FloatSlider(min=0, max=1.0, step=0.01, value=0, continuous_update=False, description= 'Alpha')
l = widgets.Dropdown(
    options=image_names,
    value=image_names[0],
    description='Image:',
    disabled=False,)

# Main part
interactive_plot = interactive(f, alpha = w, image = l);
output = interactive_plot.children[-1]
interactive_plot
```
